```python
def gauss_func(points, use_z=False):
    """ Apply gauss function over a set of points
    Args:
        points: a set of np.array points
        use_z: a bool to indicate if function will be centered on zero
    Returns:
        Two np.arrays, the first representing x values and the seconde a gauss function
    """ 
    mu = np.mean(points)
    std = np.std(points)

    if use_z:
        z_points = (points - mu)/std
        y_gauss = 1/(np.sqrt(2 * np.pi)) * np.e**(-0.5 * z_points**2)
        return z_points, y_gauss
    else:
        y_gauss = ( 1/(std * np.sqrt(2 * np.pi)) ) * np.e**( - (points-mu)**2 / std**2) 
        return points, y_gauss
    
```

```python
x_points = np.random.normal(170,10, 10000)
#x_points = np.sort(x_points)

points, y_gauss = gauss_func(x_points, use_z=False)
plt.plot(points, y_gauss,'ro')
plt.show()
```

```python
pokemon = pd.read_csv("pokemon.csv")
attack = pokemon['sp_attack']
pokemon.describe()
```

```python
plt.hist(attack, 10)
plt.show()
```

```python
def gauss_value(points, mu, std):
    y_gauss = ( 1/(std * np.sqrt(2 * np.pi)) ) * np.e**( -0.5 * (points-mu)**2 / std**2) 
    return y_gauss


mean_attack = 175.305868
std_attack = 32.353826


x_base = np.arange(0, 200)
y_gauss = gauss_value(x_base, mean_attack, std_attack)

plt.plot(attack, np.zeros(attack.shape[0]), 'ro')
plt.plot(x_base, y_gauss)

plt.show()


    

```

```python
mu_range = np.arange(50, 150, 0.5)
std_range = np.arange(30, 80, 0.5)


# forma muito burra de se fazer
best_mu = - np.inf
best_std = - np.inf
max_loglikelihood = - np.inf




for mu_t in mu_range:
    for std_t in std_range:
        y_gauss = gauss_value(attack, mu_t, std_t)
        log_likelihood = np.sum(-np.log(y_gauss))
        
        if (log_likelihood - max_loglikelihood) > 0.0001 :
            max_loglikelihood = log_likelihood
            best_mu = mu_t
            best_std = std_t

            
print("MELHOR MU: ", best_mu)
print("MELHOR STD: ", best_std)
```
